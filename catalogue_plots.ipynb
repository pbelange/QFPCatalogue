```python
import yaml
import pandas as pd
from pathlib import Path

_dict = {'config':[],'line_ID':[],'Energy':[],'I_oct':[],'Nb':[],'emittx_n':[],'emitty_n':[],'xing': []}
for config_file in list(Path('Configs/').glob(f'config_*.yaml')):
    with open(config_file) as fid:
            configuration = yaml.safe_load(fid)
            _dict['config'].append(config_file.name)
            _dict['line_ID'].append(config_file.stem[7:])
            _dict['Energy'].append(configuration['beam_energy_tot'])
            _dict['Nb'].append(configuration['beam_npart'])
            _dict['emittx_n'].append(configuration['beam_norm_emit_x'])
            _dict['emitty_n'].append(configuration['beam_norm_emit_y'])
            _dict['I_oct'].append(configuration['oct_current'])
            _dict['xing'].append(configuration['knob_settings']['on_x1'])
            assert float(configuration['knob_settings']['on_x1'])== float(configuration['knob_settings']['on_x5'])

config_df = pd.DataFrame(_dict).sort_values(by='line_ID').reset_index(drop=True)
display(config_df)
```
